# Faithfulness

```python
import pandas as pd
import re

def extract_faithfulness_score(response):
    # Check if the response is a string
    if isinstance(response, str):
        # Use regex to find the score
        match = re.search(r'\b(?:Faithfulness Rating|Faithfulness score|I rate the faithfulness|I give the report a faithfulness rating of)\s*:\s*(\d)', response, re.IGNORECASE)
        if match:
            return int(match.group(1))
    # Return NaN if the response is not a string or score is not found
    return pd.NA

def append_faithfulness_score(input_csv, output_csv):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(input_csv)
    
    # Apply the extraction function to each row in the "LLM Faithfulness Raw Response" column
    df['LLM Faithfulness Score'] = df['LLM Faithfulness Raw Response'].apply(extract_faithfulness_score)
    
    # Save the DataFrame to a new CSV file
    df.to_csv(output_csv, index=False)
    print(f"Faithfulness scores have been appended and saved to '{output_csv}'.")

# Example usage:
input_csv = 'updated_dataframe.csv'  # Replace with your input file path
output_csv = 'parsed_data.csv'  # Replace with your desired output file path
append_faithfulness_score(input_csv, output_csv)
```

# Coverage

```python
import pandas as pd
import re

def extract_coverage_score(response):
    # Check if the response is a string
    if isinstance(response, str):
        # Use regex to find the coverage score
        match = re.search(r'\b(?:Coverage Rating|coverage score|I rate the coverage|I give the coverage a score of)\s*:\s*(\d)', response, re.IGNORECASE)
        if match:
            return int(match.group(1))
    # Return NaN if the response is not a string or score is not found
    return pd.NA

def append_coverage_score(input_csv, output_csv):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(input_csv, sep=';')
    
    # Apply the extraction function to each row in the "LLM Coverage Raw Response" column
    df['LLM Coverage Score'] = df['LLM Coverage Raw Response'].apply(extract_coverage_score)
    
    # Save the DataFrame to a new CSV file
    df.to_csv(output_csv, index=False)
    print(f"Coverage scores have been appended and saved to '{output_csv}'.")

# Example usage:
input_csv = 'ParsedDataCoverage.csv'  # Replace with your input file path
output_csv = 'parsed_data2.csv'  # Replace with your desired output file path
append_coverage_score(input_csv, output_csv)
```
