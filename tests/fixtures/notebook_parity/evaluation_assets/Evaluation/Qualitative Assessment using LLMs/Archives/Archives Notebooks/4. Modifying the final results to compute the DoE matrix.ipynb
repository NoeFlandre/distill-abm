```python
import pandas as pd

# Load the CSV file
file_path = 'Final Results/Sheet to compute DOE.csv'  # Adjust this path if necessary
df = pd.read_csv(file_path, sep=';')

# Define the conditions
conditions = {
    "Summary": "BART",
    "LLM": "Claude",
    "Role": "Yes",
    "Example": "Yes",
    "Insight": "Yes"
}

# Apply the conditions
for column, value in conditions.items():
    df[column] = df[column].apply(lambda x: 1 if x == value else -1)

# Sort the DataFrame by the columns of interest
df = df.sort_values(by=["Summary", "LLM", "Role", "Example", "Insight"])

# Save the modified and sorted DataFrame back to a CSV file
output_path = 'Final Results/Sorted_Modified_Sheet_to_compute_DOE.csv'
df.to_csv(output_path, index=False)

print(f"Sorted and modified CSV saved to {output_path}")
```
