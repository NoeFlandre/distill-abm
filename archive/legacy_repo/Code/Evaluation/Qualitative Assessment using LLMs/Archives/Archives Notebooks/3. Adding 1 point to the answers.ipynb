```python
import pandas as pd

def modify_scores(input_file, output_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(input_file, sep=';')
    
    # Define the columns to be modified
    columns_to_modify = ["LLM Faithfulness Score", "LLM Coverage Score"]
    
    # Function to increment the score by 1 if it's different from 5
    def increment_score(score):
        if score != 5:
            return score + 1
        return score

    # Create new columns for the modified scores
    for column in columns_to_modify:
        new_column = "Final " + column
        df[new_column] = df[column].apply(increment_score)
    
    # Write the modified DataFrame to a new CSV file
    df.to_csv(output_file, index=False, sep=';')

# Example usage
input_file = 'parsed_data2.csv'
output_file = 'parsed_data2_with_final_scores.csv'
modify_scores(input_file, output_file)
```
