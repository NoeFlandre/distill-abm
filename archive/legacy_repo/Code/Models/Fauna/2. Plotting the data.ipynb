# Percentage of Green Patches

The percentage of landscape patches that are currently vegetated (i.e., green), indicating the availability of fodder (grass) for the prey animals.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'reduced_version.csv'

# Load the CSV file again, specifying the delimiter as ';' to correctly parse the columns
data_corrected = pd.read_csv(file_path, delimiter=';')

# Extract and display all columns containing exactly "( ( count patches with [ pcolor = green ] ) / ( count patches ) ) * 100"
green_patch_columns = [col for col in data_corrected.columns if "( ( count patches with [ pcolor = green ] ) / ( count patches ) ) * 100" in col]

# Plot all columns containing "( ( count patches with [ pcolor = green ] ) / ( count patches ) ) * 100"
plt.figure(figsize=(15, 10))

for column in green_patch_columns:
    plt.plot(data_corrected[column], label=column)

plt.xlabel('Steps')
plt.ylabel('Percentage of Green Patches')
plt.title('Percentage of Green Patches')

plt.show()


```

The percentage of green patches shows a fluctuating trend across all simulations. Initially, the percentage starts at around 32% and a bit more for some simulations. As the simulation progresses, there are sharp declines observed at regular intervals, where the percentage drops rapidly. The first declines at 400 ticks drops at around 23 to 27%, the second one at 700 ticks drops at around 29 to 32%, the third one at 1100 ticks drops at around 31 to 32%. After these drops, the percentage of green patches recovers and increases a bit before the next decline. The fourth decline at 1400 ticks drops at around 32%. From the fourth one, the percentage of green patches stabilizes at around 33% and we observe similar drops than the fourth one at around 1700 and 2200 ticks.

The model simulates an environment with grassland patches, male and female ungulates, and human hunters to explore the dynamics between these elements. The goal is to understand how hunting behaviors, herd demography, and environmental conditions interact to either accelerate or mitigate extinction events. .Key parameters set in the model include the number of initial foragers at 20, processing costs for female animals set at 3 out of 20 and 6 out of 20 for males. The maximum lifespan of animals is set at 10,220 ticks. The initial number of prey animals was 250, with a food value of 30 out of 100 for males and 27 out of 100 for females. The model employed a grass growth mechanism, with grass regrowth time set at 3 ticks and a grass proportion of 0.33. Foragers were designed to be immortal, moving with a cost of 1 energy unit per tick and having a foraging radius of 6 patches. Internal and external mortality rates for animals were set at 0.025 out of 0.2 and 0.0025 out of 0.2, respectively. The model ran for a maximum of 73,000 ticks.
