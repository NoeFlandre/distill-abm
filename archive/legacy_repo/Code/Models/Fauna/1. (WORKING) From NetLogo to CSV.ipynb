## 1.1 Loading the NetLogo file

```python
import os
import pandas as pd
import pynetlogo
import time

# Set JVM options to increase the stack size
os.environ['JAVA_OPTS'] = '-Xss4m'

# Path to the NetLogo installation directory
netlogo_home = '/Volumes/NetLogo-6.2.0/NetLogo 6.2.0'

# Create an instance of NetLogoLink
netlogo = pynetlogo.NetLogoLink(netlogo_home=netlogo_home)

# Path to the NetLogo model file
model_path = 'NetLogo/model.nlogo'

# Load the NetLogo model
netlogo.load_model(model_path)

# Define the experiment parameters
experiment_parameters = {
    "number-of-agents": 1000,
    "habit-on?": True,
    "network-type": '"watts-strogatz"',  # Wrap value in quotes
    "norms": True,
    "network-parameter": 8,
    "memory-lifetime": 5,
    "habit-threshold": 2,
    "cognitive-dissonance-threshold": 0.15,
    "p-interact": 0.3,
    "social-blindness": 0.51,
    "social-conformity": 0.28,
    "social-susceptibility": 0.2,
    "alt-health-mean-initial": 1.69,
    "alt-env-mean-initial": 1.12,
    "incumbent-initial-habit": 10,
    "justification": 0.61
}

# Define the reporters
reporters = [
    'mean-incum',
    'mean-alt',
    'incum-total-try',
    'alt-total-try',
    'choice-total',
    'mean [incum-quantity] of turtles',
    'mean [alt-quantity] of turtles',
    'mean [habit-factor-incum] of turtles',
    'mean [habit-factor-alt] of turtles',
    'mean [choice-function-deviation] of turtles',
    'mean [value-health-deviation] of turtles',
    'mean [value-env-deviation] of turtles'
]

# Create a DataFrame to store results
all_results = pd.DataFrame()

# Run the experiment for 40 repetitions
num_runs = 40
max_ticks = 73000  # Adjust as needed
interval = 50  # Adjust the interval to capture data at desired frequency

for run in range(num_runs):
    start_run_time = time.time()
    print(f"Starting run {run + 1} at {time.ctime(start_run_time)}")

    # Set the parameters
    for param, value in experiment_parameters.items():
        # Convert Python boolean to NetLogo boolean
        if isinstance(value, bool):
            value = "true" if value else "false"
        command = f'set {param} {value}'
        netlogo.command(command)

    # Setup and run the model
    netlogo.command('setup')
    print("Model setup completed. Starting simulation...")

    # Run the model for the specified number of ticks and collect data
    run_results = []
    for tick in range(0, max_ticks, interval):
        netlogo.command('go')
        # Collect data from reporters
        tick_data = {reporter: netlogo.report(reporter) for reporter in reporters}
        tick_data['tick'] = tick
        run_results.append(tick_data)
    
    # Convert run results to DataFrame and add a prefix for the run
    run_df = pd.DataFrame(run_results)
    run_df.columns = [f"{col}" for col in run_df.columns]
    
    # Append run results to all_results DataFrame horizontally
    if all_results.empty:
        all_results = run_df
    else:
        all_results = pd.concat([all_results, run_df], axis=1)

    end_run_time = time.time()
    print(f"Run {run + 1} completed successfully in {end_run_time - start_run_time:.2f} seconds.")

# Reset the index to have a proper CSV format
all_results.reset_index(drop=True, inplace=True)

# Save the DataFrame to a CSV file
output_csv_path = 'CSV/netlogo_experiment_results.csv'
all_results.to_csv(output_csv_path, index=False)

# Close the NetLogo instance when done
netlogo.kill_workspace()
print("NetLogo workspace closed successfully.")
```

## 1.2 Parsing the parameters from the GUI

```python
import re
import json

def extract_parameters(netlogo_code):
    """
    Extracts the parameters from a NetLogo code.
    
    Parameters:
    netlogo_code (str): The NetLogo code as a string.
    
    Returns:
    dict: A dictionary with parameter names as keys and their types and default values as values.
    """
    # Dictionary to store the parameters
    parameters = {
        'sliders': [],
        'switches': [],
        'monitors': [],
        'buttons': []
    }
    
    # Regular expressions to match sliders, switches, monitors, and buttons
    slider_pattern = re.compile(r'SLIDER\s+\d+\s+\d+\s+\d+\s+\d+\s+(\S+)\s+(\S+)\s+([\d.]+)\s+([\d.]+)\s+([\d.]+)\s+([\d.]+)', re.MULTILINE)
    switch_pattern = re.compile(r'SWITCH\s+\d+\s+\d+\s+\d+\s+\d+\s+(\S+)\s+(\S+)\s+(\d+)\s+(\d+)\s+(-?\d+)', re.MULTILINE)
    monitor_pattern = re.compile(r'MONITOR\s+\d+\s+\d+\s+\d+\s+\d+\s+(\S+)\s+(.*)\s+(\d+)\s+(\d+)\s+(\d+)', re.MULTILINE)
    button_pattern = re.compile(r'BUTTON\s+\d+\s+\d+\s+\d+\s+\d+\s+(NIL|\S+)\s+(NIL|\S+)\s+(T|F)\s+(T|F)\s+OBSERVER\s+(NIL|\S+)\s+(NIL|\S+)\s+(NIL|\S+)\s+(NIL|\S+)\s+(\d+)', re.MULTILINE)
    
    # Find all sliders
    for match in slider_pattern.finditer(netlogo_code):
        parameters['sliders'].append({
            'name': match.group(1),
            'min_value': float(match.group(3)),
            'max_value': float(match.group(4))
        })
    
    # Find all switches
    for match in switch_pattern.finditer(netlogo_code):
        parameters['switches'].append({
            'name': match.group(1),
            'true_value': bool(int(match.group(3))),
            'false_value': bool(int(match.group(4))),
            'default_value': bool(int(match.group(5)))
        })
    
    # Find all monitors
    for match in monitor_pattern.finditer(netlogo_code):
        parameters['monitors'].append({
            'name': match.group(1),
            'reporter': match.group(2),
            'x': int(match.group(3)),
            'y': int(match.group(4)),
            'width': int(match.group(5))
        })
    
    # Find all buttons
    for match in button_pattern.finditer(netlogo_code):
        parameters['buttons'].append({
            'display_name': match.group(1),
            'procedure': match.group(2),
            'forever': match.group(3) == 'T',
            'turtle_context': match.group(4) == 'T',
            'obs1': match.group(5),
            'obs2': match.group(6),
            'obs3': match.group(7),
            'obs4': match.group(8),
            'state': int(match.group(9))
        })
    
    # Remove empty elements
    parameters = {key: value for key, value in parameters.items() if value}
    
    return parameters

# Path to your NetLogo file
file_path = 'NetLogo/model.nlogo'

# Read the contents of the NetLogo file
with open(file_path, 'r') as file:
    netlogo_code = file.read()

# Extract the parameters
parameters = extract_parameters(netlogo_code)

# Path to save the extracted parameters
output_file_path = 'JSON/gui_parameters100.json'

# Save the extracted parameters to a JSON file
with open(output_file_path, 'w') as json_file:
    json.dump(parameters, json_file, indent=4)

# Print a confirmation message
print(f"Parameters have been extracted and saved to {output_file_path}")
```

# Adding the values from the experiment to the GUI

```python
import json

# Paths to the input JSON files
gui_parameters_path = 'JSON/gui_parameters100.json'
experiment_parameters_path = 'JSON/experiment_parameters100.json'

# Paths to the output JSON files
updated_gui_parameters_path = 'JSON/updated_gui_parameters100.json'
updated_experiment_parameters_path = 'JSON/updated_experiment_parameters100.json'

# Load the GUI parameters JSON
with open(gui_parameters_path, 'r') as gui_file:
    gui_parameters = json.load(gui_file)

# Load the experiment parameters JSON
with open(experiment_parameters_path, 'r') as experiment_file:
    experiment_parameters = json.load(experiment_file)

# Function to update parameters and ensure all required fields are present
def update_parameters(gui_section, experiment_parameters):
    for item in gui_section:
        name = item['name']
        if name in experiment_parameters:
            item['value'] = experiment_parameters[name]
            del experiment_parameters[name]

# Update each section in gui_parameters
for section in ['sliders', 'switches', 'monitors', 'buttons']:
    if section in gui_parameters:
        update_parameters(gui_parameters[section], experiment_parameters)

# Function to format the JSON with each parameter on one line
def format_json_oneline(data):
    formatted_json = "{\n"
    for section, items in data.items():
        formatted_json += f'  "{section}": [\n'
        for item in items:
            formatted_item = json.dumps(item, separators=(',', ': '))
            formatted_json += f'    {formatted_item},\n'
        if items:
            formatted_json = formatted_json[:-2]  # Remove the trailing comma and newline
        formatted_json += '\n  ],\n'
    if data:
        formatted_json = formatted_json[:-2]  # Remove the trailing comma and newline
    formatted_json += "\n}"
    return formatted_json

# Save the modified GUI parameters JSON to a new file with the custom formatting
formatted_gui_json = format_json_oneline(gui_parameters)
with open(updated_gui_parameters_path, 'w') as updated_gui_file:
    updated_gui_file.write(formatted_gui_json)
print(f"Updated GUI parameters have been saved to {updated_gui_parameters_path}")

# Save the modified experiment parameters JSON to a new file
with open(updated_experiment_parameters_path, 'w') as updated_experiment_file:
    json.dump(experiment_parameters, updated_experiment_file, indent=4)
print(f"Updated experiment parameters have been saved to {updated_experiment_parameters_path}")
```

## 1.3 Parsing the documentation

```python
import re
import json

def extract_documentation(file_path):
    """
    Extracts the documentation section from a NetLogo file.

    Parameters:
    file_path (str): The path to the NetLogo file.

    Returns:
    str: The extracted documentation section.
    """
    # Read the contents of the NetLogo file
    with open(file_path, 'r') as file:
        content = file.read()

    # Find the start and end of the documentation section
    start_pattern = re.compile(r"@#\$#@#\$#@\s*## WHAT IS IT\?")
    end_pattern = re.compile(r"@#\$#@#\$#@")

    start_match = start_pattern.search(content)

    if not start_match:
        raise ValueError("Start of the documentation section not found in the NetLogo file.")

    end_match = end_pattern.search(content, start_match.end())

    if not end_match:
        raise ValueError("End of the documentation section not found in the NetLogo file.")

    # Extract the documentation section
    documentation = content[start_match.start():end_match.start()].strip()

    return documentation

# Path to your NetLogo file
file_path = 'NetLogo/model.nlogo'

# Extract the documentation
documentation = extract_documentation(file_path)

# Prepare the data to be saved in JSON format
data = {
    'documentation': documentation
}

# Path to save the extracted documentation
output_file_path = 'JSON/documentation100.json'

# Save the extracted documentation to a JSON file
with open(output_file_path, 'w') as json_file:
    json.dump(data, json_file, indent=4)

# Print a confirmation message
print(f"Documentation has been extracted and saved to {output_file_path}")
```

# Remove urls

```python
import json
import re

def remove_urls(text):
    # Define a regex pattern to match URLs
    url_pattern = re.compile(r'http[s]?://\S+|www\.\S+')
    # Substitute URLs with an empty string
    return url_pattern.sub('', text)

def process_documentation(input_json, output_json):
    # Read the input JSON file
    with open(input_json, 'r') as f:
        data = json.load(f)
    
    # Function to recursively remove URLs from all strings in the JSON data
    def remove_urls_from_data(data):
        if isinstance(data, dict):
            return {key: remove_urls_from_data(value) for key, value in data.items()}
        elif isinstance(data, list):
            return [remove_urls_from_data(element) for element in data]
        elif isinstance(data, str):
            return remove_urls(data)
        else:
            return data

    # Process the JSON data to remove URLs
    updated_data = remove_urls_from_data(data)
    
    # Save the updated JSON data to a new file
    with open(output_json, 'w') as f:
        json.dump(updated_data, f, indent=4)

if __name__ == "__main__":
    input_json = 'JSON/documentation100.json'  # Replace with your input JSON file path
    output_json = 'JSON/cleaneddocumentation100.json'  # Replace with your desired output JSON file path
    process_documentation(input_json, output_json)
```

# Remove the default elements

```python
import json

# Define default NetLogo documentation elements
default_elements = {
    "## WHAT IS IT?": "(a general understanding of what the model is trying to show or explain)",
    "## HOW IT WORKS": "(what rules the agents use to create the overall behavior of the model)",
    "## HOW TO USE IT": "(how to use the model, including a description of each of the items in the Interface tab)",
    "## THINGS TO NOTICE": "(suggested things for the user to notice while running the model)",
    "## THINGS TO TRY": "(suggested things for the user to try to do (move sliders, switches, etc.) with the model)",
    "## EXTENDING THE MODEL": "(suggested things to add or change in the Code tab to make the model more complicated, detailed, accurate, etc.)",
    "## NETLOGO FEATURES": "(interesting or unusual features of NetLogo that the model uses, particularly in the Code tab; or where workarounds were needed for missing features)",
    "## RELATED MODELS": "(models in the NetLogo Models Library and elsewhere which are of related interest)",
    "## CREDITS AND REFERENCES": "(a reference to the model's URL on the web if it has one, as well as any other necessary credits, citations, and links)"
}

def remove_default_elements(text):
    # Use a list to store cleaned sections
    cleaned_sections = []
    # Split the documentation into sections
    sections = text.split("\n## ")
    
    for section in sections:
        if not section.strip():
            continue  # Skip empty sections

        # Identify the header and content
        if "\n\n" in section:
            header, content = section.split("\n\n", 1)
            header = "## " + header.strip()
        else:
            header, content = "## " + section.strip(), ""
        
        # Check if the content matches the default
        if header in default_elements and default_elements[header] in content.strip():
            continue  # Skip the default section
        
        cleaned_sections.append(header + "\n\n" + content.strip())
    
    return "\n".join(cleaned_sections).strip()

def process_documentation(input_json, output_json):
    # Read the input JSON file
    with open(input_json, 'r') as f:
        data = json.load(f)
    
    # Process the JSON data to remove default elements
    if "documentation" in data:
        documentation = data["documentation"]
        cleaned_documentation = remove_default_elements(documentation)
        data["documentation"] = cleaned_documentation
    
    # Save the updated JSON data to a new file
    with open(output_json, 'w') as f:
        json.dump(data, f, indent=4)

if __name__ == "__main__":
    input_json = 'JSON/cleaneddocumentation100.json'  # Replace with your input JSON file path
    output_json = 'JSON/documentationWithoutDefault100.json'  # Replace with your desired output JSON file path
    process_documentation(input_json, output_json)
```

# Cleaning the headers

```python
import json
import re

def clean_json_content(input_json, output_txt):
    # Load the JSON data
    with open(input_json, 'r') as file:
        data = json.load(file)
    
    # Extract the documentation content
    documentation = data.get("documentation", "")

    # Remove the unwanted section "## @#$#@#$#@"
    cleaned_content = re.sub(r'## @#\$#@#\$#@\n\n\n', '', documentation)

    # Remove all headers that start with "##"
    cleaned_content = re.sub(r'## .*?\n', '', cleaned_content)

    # Save the cleaned content to a text file
    with open(output_txt, 'w') as file:
        file.write(cleaned_content)

# Usage example
input_json = 'JSON/documentationWithoutDefault100.json'
output_txt = 'TXT/finalDocumentation100.txt'
clean_json_content(input_json, output_txt)
```

## 1.4 Parsing the code section

```python
import re

def extract_code(file_path):
    """
    Extracts the code section from a NetLogo file.

    Parameters:
    file_path (str): The path to the NetLogo file.

    Returns:
    str: The extracted code section.
    """
    # Read the contents of the NetLogo file
    with open(file_path, 'r') as file:
        content = file.read()

    # Find the start of the code section
    start_pattern = re.compile(r"globals\s*\[")
    end_pattern = re.compile(r"@#\$#@#\$#@")

    start_match = start_pattern.search(content)
    end_match = end_pattern.search(content, start_match.end() if start_match else 0)

    if not start_match or not end_match:
        raise ValueError("Code section not found in the NetLogo file.")

    # Extract the code section
    code_section = content[start_match.start():end_match.start()].strip()

    return code_section

# Path to your NetLogo file
file_path = 'NetLogo/model.nlogo'
# Path to save the extracted code
output_code_path = 'TXT/extracted_code100.txt'

# Extract the code
code_section = extract_code(file_path)

# Save the extracted code to a text file
with open(output_code_path, 'w') as file:
    file.write(code_section)

# Print confirmation
print(f"Extracted code section has been saved to {output_code_path}")
```
