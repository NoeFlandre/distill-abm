```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'reduced3.csv'
data_corrected = pd.read_csv(file_path, delimiter=';')
```

```python
def plot_columns(data, column_pattern, y_label, title, exclude_pattern=None):
    # Extract columns containing the specified pattern and not containing the exclude pattern
    if exclude_pattern:
        columns = [col for col in data.columns if column_pattern in col and exclude_pattern not in col]
    else:
        columns = [col for col in data.columns if column_pattern in col]

    # Inspect the data for any non-numeric values or other anomalies
    for column in columns:
        try:
            # Convert column to numeric, forcing any errors to NaN
            data[column] = pd.to_numeric(data[column], errors='coerce')
        except Exception as e:
            print(f"Error processing column {column}: {e}")

    # Plot all columns containing the specified pattern
    plt.figure(figsize=(15, 10))

    for column in columns:
        plt.plot(data[column].dropna().reset_index(drop=True), label=column)

    plt.xlabel('Steps')
    plt.ylabel(y_label)
    plt.title(title)
    
    # Clean up y-axis scale by setting a fixed number of ticks
    plt.gca().yaxis.set_major_locator(plt.MaxNLocator(10))

    plt.show()
```

# Mean Initial Household Risk Attitude

The "Mean Initial Household Risk Attitude" metric represents the average initial risk attitude of all social learning agents, used to store and report their initial disposition to deviate from the recommended number of livestock.

```python
plot_columns(data_corrected, "mean [household-risk-att-init] of households", 'Mean Household Initial Risk Attitude', 'Mean Household Initial Risk Attitude')
```

At the start of the simulation, the mean initial household risk attitudes are distributed across a range of values, indicating varying degrees of risk propensity among the households. The values are evenly spread from approximately -0.15 to 0.15, reflecting diverse initial attitudes towards risk-taking.

As time progresses, the mean initial risk attitudes for each household remain constant. This stability indicates that there are no changes in the initial risk attitudes over the simulation period, suggesting that the initial conditions set for the households are maintained throughout the 100 time steps.
