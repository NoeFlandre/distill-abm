#  Remove www

```python
import pandas as pd
import re

# Read the CSV file
input_file = 'CSV/Refining/ParagraphBERTR1Janus.csv'
output_file = 'CSV/Refining/wwwRemovedR1Janus.csv'

# Load the data
df = pd.read_csv(input_file)

# Define the columns to process
columns_to_process = ['Summary (BART) Reduced', 'Summary (BERT) Reduced']

def remove_sentences_with_www(text):
    # Split the text into paragraphs first
    paragraphs = text.split('\n')
    new_paragraphs = []
    
    for paragraph in paragraphs:
        # Split the paragraph into sentences using regex to handle various punctuation and whitespace cases
        sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', paragraph)
        filtered_sentences = [sentence for sentence in sentences if 'www.' not in sentence]
        new_paragraph = ' '.join(filtered_sentences).strip()
        new_paragraphs.append(new_paragraph)
    
    # Join the paragraphs back together with newlines
    return '\n'.join(new_paragraphs).strip()

# Apply the function to each specified column
for column in columns_to_process:
    if column in df.columns:
        df[column] = df[column].apply(lambda x: remove_sentences_with_www(str(x)))

# Write the modified data to a new CSV file
df.to_csv(output_file, index=False)

print(f"Processed CSV saved as {output_file}")
```

# Hyphen After Punctuation

```python
import pandas as pd
import re

def remove_hyphens_after_punctuation(text):
    # Remove hyphens that are preceded by punctuation and spaces
    text = re.sub(r'([.,!?;:]\s*)-\s*', r'\1', text)
    return text

def main(input_csv, output_csv):
    # Read the input CSV
    df = pd.read_csv(input_csv)
    
    # Process the "Summary (BERT) Reduced" column
    if 'Summary (BERT) Reduced' in df.columns:
        df['Summary (BERT) Reduced'] = df['Summary (BERT) Reduced'].apply(lambda x: remove_hyphens_after_punctuation(x) if pd.notnull(x) else x)
    
    # Save the modified dataframe to a new CSV file
    df.to_csv(output_csv, index=False)

if __name__ == "__main__":
    input_csv = 'CSV/Refining/wwwRemovedR1Janus.csv'  # Replace with your input CSV file path
    output_csv = 'CSV/Refining/HyphenFixedR1Janus.csv'  # Replace with your desired output CSV file path
    main(input_csv, output_csv)
```

# Punctuation at the beginning of a sentence

```python
import pandas as pd
import re

def remove_unnecessary_punctuation(text):
    # Remove punctuation at the beginning of sentences
    text = re.sub(r'([.!?]\s*)[,;:]\s*', r'\1', text)
    return text

def main(input_csv, output_csv):
    # Read the input CSV
    df = pd.read_csv(input_csv)
    
    # Process the "Summary (BERT) Reduced" column
    if 'Summary (BERT) Reduced' in df.columns:
        df['Summary (BERT) Reduced'] = df['Summary (BERT) Reduced'].apply(lambda x: remove_unnecessary_punctuation(x) if pd.notnull(x) else x)
    
    # Save the modified dataframe to a new CSV file
    df.to_csv(output_csv, index=False)

if __name__ == "__main__":
    input_csv = 'CSV/Refining/HyphenFixedR1Janus.csv'  # Replace with your input CSV file path
    output_csv = 'CSV/Refining/PunctuationBeginningSentenceFixedR1Janus.csv'  # Replace with your desired output CSV file path
    main(input_csv, output_csv)
```

# Spaces in Parenthesis

```python
import pandas as pd
import re

def remove_unnecessary_spaces_in_parentheses(text):
    # Remove unnecessary spaces inside parentheses
    text = re.sub(r'\(\s*', '(', text)
    text = re.sub(r'\s*\)', ')', text)
    return text

def main(input_csv, output_csv):
    # Read the input CSV
    df = pd.read_csv(input_csv)
    
    # Process the "Summary (BERT) Reduced" column
    if 'Summary (BERT) Reduced' in df.columns:
        df['Summary (BERT) Reduced'] = df['Summary (BERT) Reduced'].apply(lambda x: remove_unnecessary_spaces_in_parentheses(x) if pd.notnull(x) else x)
    
    # Save the modified dataframe to a new CSV file
    df.to_csv(output_csv, index=False)

if __name__ == "__main__":
    input_csv = 'CSV/Refining/PunctuationBeginningSentenceFixedR1Janus.csv'  # Replace with your input CSV file path
    output_csv = 'CSV/Refining/SpacesParenthesisR1Janus.csv'  # Replace with your desired output CSV file path
    main(input_csv, output_csv)
```

# Space before dot

```python
import pandas as pd
import re

def remove_space_before_dot(text):
    # Remove space before a dot
    text = re.sub(r'\s+\.', '.', text)
    return text

def main(input_csv, output_csv):
    # Read the input CSV
    df = pd.read_csv(input_csv)
    
    # Process the "Summary (BART) Reduced" column
    if 'Summary (BART) Reduced' in df.columns:
        df['Summary (BART) Reduced'] = df['Summary (BART) Reduced'].apply(lambda x: remove_space_before_dot(x) if pd.notnull(x) else x)
    
    # Save the modified dataframe to a new CSV file
    df.to_csv(output_csv, index=False)

if __name__ == "__main__":
    input_csv = 'CSV/Refining/SpacesParenthesisR1Janus.csv'  # Replace with your input CSV file path
    output_csv = 'CSV/Refining/SpacesBeforeDotR1Janus.csv'  # Replace with your desired output CSV file path
    main(input_csv, output_csv)
```

# Upper case

```python
import pandas as pd
import re

def capitalize_sentences(text):
    # Define a function to capitalize the first letter of each sentence
    def capitalize(match):
        return match.group(1) + match.group(2).upper()
    
    # Capitalize the first letter of the text
    text = text[0].upper() + text[1:] if text else text
    
    # Use regex to find sentence endings followed by a lowercase letter
    text = re.sub(r'([.!?]\s+)([a-z])', capitalize, text)
    
    # Use regex to find paragraph boundaries followed by a lowercase letter
    text = re.sub(r'(\n\s*)([a-z])', capitalize, text)
    
    return text

def main(input_csv, output_csv):
    # Read the input CSV
    df = pd.read_csv(input_csv)
    
    # Process the "Summary (BERT) Reduced" column
    if 'Summary (BERT) Reduced' in df.columns:
        df['Summary (BERT) Reduced'] = df['Summary (BERT) Reduced'].apply(lambda x: capitalize_sentences(x) if pd.notnull(x) else x)
    
    # Save the modified dataframe to a new CSV file
    df.to_csv(output_csv, index=False)

if __name__ == "__main__":
    input_csv = 'CSV/Refining/SpacesBeforeDotR1Janus.csv'  # Replace with your input CSV file path
    output_csv = 'CSV/Refining/UpperCaseR1Janus.csv'  # Replace with your desired output CSV file path
    main(input_csv, output_csv)
```

# Unicode

```python
import csv
from unidecode import unidecode

def clean_non_unicode(input_file, output_file):
    def remove_non_unicode(text):
        return unidecode(text)
    
    with open(input_file, mode='r', encoding='utf-8') as infile, open(output_file, mode='w', encoding='utf-8', newline='') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)
        
        for row in reader:
            cleaned_row = [remove_non_unicode(cell) for cell in row]
            writer.writerow(cleaned_row)

# Example usage
input_csv = 'CSV/Refining/UpperCaseR1Janus.csv'
output_csv = 'CSV/Refining/UnicodeR1Janus.csv'
clean_non_unicode(input_csv, output_csv)

print(f"Cleaned CSV saved as {output_csv}")
```
