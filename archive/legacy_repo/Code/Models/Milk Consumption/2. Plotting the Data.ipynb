```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'CSV/reducedmilk.csv'
data_corrected = pd.read_csv(file_path, delimiter=';')
```

```python
def plot_columns(data, column_pattern, y_label, title, exclude_pattern=None):
    # Extract columns containing the specified pattern and not containing the exclude pattern
    if exclude_pattern:
        columns = [col for col in data.columns if column_pattern in col and exclude_pattern not in col]
    else:
        columns = [col for col in data.columns if column_pattern in col]

    # Inspect the data for any non-numeric values or other anomalies
    for column in columns:
        try:
            # Convert column to numeric, forcing any errors to NaN
            data[column] = pd.to_numeric(data[column], errors='coerce')
        except Exception as e:
            print(f"Error processing column {column}: {e}")

    # Plot all columns containing the specified pattern
    plt.figure(figsize=(15, 10))

    for column in columns:
        plt.plot(data[column].dropna().reset_index(drop=True), label=column)

    plt.xlabel('Steps')
    plt.ylabel(y_label)
    plt.title(title)
    
    # Clean up y-axis scale by setting a fixed number of ticks
    plt.gca().yaxis.set_major_locator(plt.MaxNLocator(10))

    plt.show()
```

# Mean Whole Milk Consumption

The average weekly consumption of whole milk per agent.

```python
plot_columns(data_corrected, "mean-incum", 'Mean Whole Milk Consumption', 'Mean Whole Milk Consumption')
```

At the start of the simulation, whole milk consumption is very high. This mirrors historical data where whole milk was the dominant choice among the British public.

In the early stages (up to about time step 10), there is a rapid decline in whole milk consumption. This phase likely represents early adopters switching to skimmed and semi-skimmed milk due to health benefits, environmental concerns, or the influence of social factors.

After times steps 10, the rate of decline slows down but continues steadily. This could indicate that a major percentage of the population has influenced the remaining minority to switch to alternative milk options.

The model focuses on individual preferences and decision factors to explore how environmental and health values, habits, and social influences affect milk choice decisions. The model employs both a threshold-based and a probability-based approach to determine how agents decide on milk choices. The simulation starts with agents who initially prefer whole milk.
